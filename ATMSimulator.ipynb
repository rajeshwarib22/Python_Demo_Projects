```python
# ATM 1. available balance 2. deposit 3.withdraw 4. exit

#set a initial bank balance amount
bal_amt= 1000.00

#to run the code in a loop until the user enters valid value for grading
while True:
      #propmt for user to select which operation to perform
      print("\n\nWelcome to ATM\n"
            "1. Available balance \n"
            "2. Deposit amount \n"
            "3. withdraw amount \n"
            "4. Exit\n")
      #to take choice from user which operation he wants to perform
      choice = input("\nPlease enter a choice :  ")

      # to display account balance
      if choice=="1":
        print("\n Your available balance is: $"+str(bal_amt) )
      # to deposit amount in the account
      elif choice=="2":
        deposit = float(input("\nPlease enter the amount to deposit : "))
        if deposit > 0:
          bal_amt += deposit
          print(f"\nThe updated balance is: ${bal_amt:.2f} \n")
        else:
          print("\nPlease enter valid input..!!!\n")
      # to withdraw amount from the account
      elif choice=="3":
        withdraw = float(input("\nPlease enter the amount to deposit : "))
        if withdraw > 0 and  bal_amt >= withdraw :    # condition to not accept withdraw amount greater than available balance
          bal_amt -= withdraw
          print(f"\nThe updated balance is: ${bal_amt:.2f} \n")
        else:
          print("\nPlease enter valid input...!!!!")
      #to exit from the operations
      elif choice=="4":
        print("\nThank you...  Have a nice day!!")
        break
      else:
        print("\nInvalid entry.....Please enter a valid choice")
```
