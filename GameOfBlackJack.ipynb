```python
# importing python library
import random

# creating deck
ranks = ['2','3','4','5','6','7','8','9','10','Jack','Queen','King','Ace']
suits = ['Hearts','Diamonds','Club','Spade']

deck = [(rank,suit) for rank in ranks for suit in suits]
deck *= 6      # creating 6 decks
#print deck
# deck

# shuffle cards
def shuffle_deck(deck):
  random.shuffle(deck)

# to deal cards
def deal_card(deck):
  return  deck.pop(0)

# function to calculate hand value
def calculate_hand_value(hand):
  hand_value = 0
  ace_counter = 0
  for card in hand:
    rank = card[0]
    if rank in ['Jack','Queen','King']:
      hand_value += 10
    elif rank == 'Ace' :
      hand_value += 11
      ace_counter +=1
    else:
      hand_value += int(rank)

  if ace_counter > 0 and hand_value >21:
    hand_value -= 10

  return hand_value

# funtion to create game logic
def main():
  player_score = 0
  dealer_score = 0
  # to give prompt to user to play the game
  play_again = input('Would you like to play a game of cards (Yes/No)? ').lower()

  while play_again == 'yes':
    shuffle_deck(deck)
    print("\nLets play a round of blackjack!")

    player_hand = [deal_card(deck), deal_card(deck)]

    #display the player's hand
    while True:
      print(f"Your hand: {', '.join([f'{card[0]} of {card[1]}' for card in player_hand])}")
      player_hand_value = calculate_hand_value(player_hand)
      print(f'Total value of your hand: {player_hand_value}')

      if player_hand_value > 21:
        print('Bust! You lose this round!')
        break

      action = input("Do you want to 'hit' or 'stand'? ").lower()
      if action == 'hit':
        player_hand.append(deal_card(deck))
      elif action == 'stand':
        break
      else:
        print("Invalid input. Please enter 'hit' or 'stand'.")

    dealer_hand = [deal_card(deck), deal_card(deck)]

    #creating dealer and win condition for card game
    while calculate_hand_value(dealer_hand) < 17:
      dealer_hand.append(deal_card(deck))

    print(f"Dealer's hand: {', '.join([f'{card[0]} of {card[1]}' for card in dealer_hand])}")
    dealer_hand_value = calculate_hand_value(dealer_hand)
    print(f"Total value of dealer's hand: {dealer_hand_value}\n")

    if dealer_hand_value > 21 or (player_hand_value <= 21 and player_hand_value > dealer_hand_value):
      print("You win this round!\n")
      player_score += 1
    else:
      print("Dealer wins this round!\n")
      dealer_score += 1

    while True:
      continue_playing = input("Do you want to play another round? (Yes/No): ").lower()
      if continue_playing == 'yes':
        break
      elif continue_playing == 'no':
        play_again = 'no'
        break
      else:
        print("Invalid option! Please enter Yes / No ")

  print(f"Game Over! Your final score: Player won {player_score} rounds, Dealer won {dealer_score} rounds.")
  print('Thanks for playing!')


main()
```
