```python
# HW#2  - python program that calculates and assigns grades based on a student's test score usinf if and else statements

#to run the code in a loop until the user enters valid value for grading
while True:
  try:
    #take input from the user
    print("Enter student test scores(range 0-100) : ")
    test_score = float(input())

    #to grade the test score of the student
    if(test_score > 100 or test_score < 0):          #if the user enters a score that is more than 100 or a negative value
      raise ValueError
    if (test_score >= 90 and test_score <= 100):     #if the test score is between range 90-100 assign grade A
      print("Your grade is A")
    elif (test_score >= 80 and test_score <= 89):    #if the test score is between range 80-89 assign grade B
      print("your grade is B")
    elif (test_score >= 70 and test_score <= 79):    #if the test score is between range 70-79 assign grade C
      print("your grade is C")
    elif (test_score >= 60 and test_score <= 69):    #if the test score is between range 60-69 assign grade D
      print("your grade is D")
    elif (test_score < 60 and test_score >= 0):      #if the test score is below 60 assign grade F
      print("your grade is F")

    break
 #exception if any invalid value is entered
  except ValueError:
        print("Oops!  Please enter a valid number.  Try again...!!!\n")

```
