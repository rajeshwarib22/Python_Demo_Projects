```python
# Part II - Short Project Python Response - 70 Points
# Write a short program:
# Develop a Task Tracker program in Python. Consider the following specifications:
# •	Use lists or tuples to store tasks with attributes like task name, due date, and priority level.
# •	Implement functions to add new tasks, display all tasks, and delete a task based on user input.
# •	Allow users to mark tasks as completed, and update the priority level of tasks.
# •	Display the tasks in a tabular format.



# this funtion is used to add new task in th e tasks list with the help of append() function
def add_task(tasks, name, due_date, priority, status = False):
  tasks.append({"name" : name, "due_date" : due_date, "priority" : priority, "status" :status})


# this funtion is used to display all the tasks present in the tasks list
def display_task(tasks):
  if not tasks:   # used to check if the tasks list is empty
    print("\n\nNo tasks to display")
  else:
    print("\nTask Tracker")
    print("\n\n")
    print("{:<30} {:<25} {:<15} {:<10}".format("Task Name", "Due Date", "Priority", "Status"))
    # if conndition is used to assign status to "completed" or "pending"
    for task in tasks:
            print("{:<30} {:<25} {:<15} {:<10}".format(task["name"], task["due_date"], task["priority"], "Completed" if task["status"] else "Pending"))
  print("\n\n")


#this function is used to delete the tasks from the task list by taking task name as input and uses remove() to remove the task entry
def delete_task(tasks, task_name):
  for task in tasks:
        if task["name"].lower() == task_name.lower():
            tasks.remove(task)
            print(f"Task '{task_name}' deleted.")
            return
  print(f"Task '{task_name}' not found.")

  print("\n\n")

#this funtion is used to mark the status as completed for a task
def task_status(tasks, task_name):
  for task in tasks:
        if task["name"].lower() == task_name.lower():
            # if condition is used to change the status value from False to True
            task["status"] = True
            print(f"Task '{task_name}' marked as completed.")
            return
  print(f"Task '{task_name}' not found.")
  print("\n\n")


# this funtion is used to update the priority of the tasks from the tasks list
def update_prority(tasks, task_name, new_priority):
  for task in tasks:
        if task["name"].lower() == task_name.lower():
          # if condition is used to replace the old value with new value
          task["priority"] = new_priority
          print(f"Priority of task '{task_name}' updated to {new_priority}.")
          return
  print(f"Task '{task_name}' not found.")
  print("\n\n")

def main():

  # intialize tasks list with empty list
  tasks = []

  # while is used to iterate the option menu for user if they want to perform more than 1 operation
  while True:

    #create a menu for Task Tracker
    print("*********Welcome to Task Tracker**********")
    print("\n\nTask Tracker Menu: ")
    print("\n1. Add Task")
    print("\n2. Display All Taks")
    print("\n3. Delete Task")
    print("\n4. Mark Task as Completed")
    print("\n5. Update Task Priority")
    print("\n6. Exit ")

    #take choice as input from user to perform the operation from the menu
    choice = input("\n\nEnter your choice (1-6) : ")

    #this choice is used to add a new Task in the Task Tracker
    if choice == '1':
      name = input("\nEnter Task Name : ")
      due_date = input("\nEnter Task Due Date (MM/DD/YYYY): ")
      priority = input("\nEnter Task Priority(High / Low / Medium) : ")
      add_task(tasks, name, due_date, priority)   # to call add_task function

    #this choice is used to display the Tasks in the Task Tracker
    elif choice == '2':
      display_task(tasks)    # to call display_task function

    #this choice is used to delete a Task in the Task Tracker
    elif choice == '3':
      task_name = input("\nEnter Task Name : ")
      delete_task(tasks, task_name)   # to call delete_task function

    #this choice is used to update the Task to completed from pending in the Task Tracker
    elif choice == '4':
      task_name = input("Enter task name to mark as completed : ")
      task_status(tasks, task_name)   # to call task_status function


    #this choice is used to update a prority(High / Low / Medium) for Task in the Task Tracker
    elif choice == '5':
      task_name = input("\nEnter task name to update priority : ")
      new_priority = input("\nEnter new priority(High / Low / Medium) : ")
      update_prority(tasks, task_name, new_priority)  # to call update_prority function

   #this choice is used to exit the Task Tracker
    elif choice == '6':
      print("\nExiting the Task Tacker. Have a Good Day!!!")
      break

    #this is a default choice used to promt the user to enter correct value from the Task Tracker menu
    else:
      print("\nPlease enter a correct option between 1-6")





# to call main function
main()
```
